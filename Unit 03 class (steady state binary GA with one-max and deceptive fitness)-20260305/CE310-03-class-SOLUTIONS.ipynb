*Binary Steady-State Genetic Algorithm*

In this class you will to implement, step by step, a binary genetic algorithms. The algorithm will use the steady state approach, i.e., it does *not* use separate populations representing the indivuals in the current generation and those of the new generation. Instead, *new individuals created via genetic operations will immediately be inserted in the population*. 

To get the algorithm working you will need to implement all the basic operations the GA and its key data structures. The operations will be: initialisation, fitness evaluation, selection, mutation, crossover, and selection of individual to be replaced. The data structures will be: a population and a place where to store the fitness values of the individuals in the population.

Once ready, you will use this GA to find the maxima of the following two fitness functions:

```python
def onemax_fitness(genome):
    return sum(genome) # Fitness is simply the count of 1s.

def deceptive_fitness(genome):
    u = sum(genome)       # Unitation (number of 1s)
    L = len(genome)       # Length of genome
    
    if L - u <= 2: # Near the global optimum
        return u
    else:          # Near a less good optimum
        return L - u - 1 

```

````python
# Set up fitnesses

def onemax_fitness(genome):
    return sum(genome)

def deceptive_fitness(genome):
    u = sum(genome)
    L = len(genome)
    if L - u <= 2:
        return u
    else:
        return L - u - 1

print('onemax')
print(onemax_fitness([0,0,0,0,0,0,0]))
print(onemax_fitness([1,0,0,0,0,0,0]))
print(onemax_fitness([1,1,0,0,0,0,0]))
print(onemax_fitness([1,1,1,0,0,0,0]))
print(onemax_fitness([1,1,1,1,0,0,0]))
print(onemax_fitness([1,1,1,1,1,0,0]))
print(onemax_fitness([1,1,1,1,1,1,0]))
print(onemax_fitness([1,1,1,1,1,1,1]))

print('deceptive')
print(deceptive_fitness([0,0,0,0,0,0,0]))
print(deceptive_fitness([1,0,0,0,0,0,0]))
print(deceptive_fitness([1,1,0,0,0,0,0]))
print(deceptive_fitness([1,1,1,0,0,0,0]))
print(deceptive_fitness([1,1,1,1,0,0,0]))
print(deceptive_fitness([1,1,1,1,1,0,0]))
print(deceptive_fitness([1,1,1,1,1,1,0]))
print(deceptive_fitness([1,1,1,1,1,1,1]))
````

````python
# Tasks 1-3
from random import randint
from pprint import pprint

def random_individual(N):
    return [randint(0, 1) for _ in range(N)]

def random_population(M, N):
    return [random_individual(N) for _ in range(M)]

def eval_fitness(population, fitness_func):
    return [fitness_func(ind) for ind in population]

# Initialize Global Parameters
N = 10          # Genome length
M = 5          # Population size
mutation_rate = 1 / N

population = random_population(M, N)
fitnesses = eval_fitness(population, onemax_fitness)
pprint(population)
print(fitnesses)
````

````python
# Tasks 4 & 7
from random import sample

def tournament():
    i, j = sample(range(M), 2)
    return i if fitnesses[i] >= fitnesses[j] else j

def negative_tournament():
    i, j = sample(range(M), 2)
    # Returns the index of the weaker individual to be replaced
    return i if fitnesses[i] <= fitnesses[j] else j

print('tourn')
for _ in range(5):
    print(tournament())

print('neg tourn')
for _ in range(5):
    print(negative_tournament())
````

````python
# Tasks 5 & 6
from random import random

def mutate():
    idx = tournament()
    parent = population[idx]
    # Create offspring by flipping bits
    offspring = [(1 - bit) if random() < mutation_rate else bit for bit in parent]
    return offspring

def crossover():
    idx1 = tournament()
    idx2 = tournament()
    parent1, parent2 = population[idx1], population[idx2]
    
    # One-point crossover
    point = randint(1, N - 1)
    offspring = parent1[:point] + parent2[point:]
    return offspring

print('XO')
for _ in range(5):
    print(crossover())

print('MUT')
for _ in range(5):
    print(mutate())
````

````python
# Tasks 8 & 9

crossover_prob = 0.5
iterations = 500
M = 50
N = 20
mutation_rate = 1 / N
fitness_function = deceptive_fitness

population = random_population(M, N)
fitnesses = eval_fitness(population, fitness_function)

# Track the best individual
best_idx = fitnesses.index(max(fitnesses))
best_genome = population[best_idx][:]
best_fit = fitnesses[best_idx]

for i in range(iterations):
    if random() < crossover_prob:
        offspring = crossover()
    else:
        offspring = mutate()
        
    f = fitness_function(offspring)
    
    k = negative_tournament()
    population[k] = offspring
    fitnesses[k] = f
    
    if f > best_fit:
        best_fit = f
        best_genome = offspring[:]
        
pprint(population)
print(best_fit)
````
