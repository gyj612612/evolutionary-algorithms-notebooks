   
2. Study the following simulator of knapsack filling process which loads the items in the order they are provided until the maximum capacity is exceeded, at which points it returns the items loaded in the knapsack and value of such items.

```python
def fill_knapsack(items, capacity = 7):
    knapsack = []
    value = 0
    remaining_capacity = capacity
    for item_name, item_size, item_value in items:
        if item_size > remaining_capacity: 
            break
        value += item_value
        remaining_capacity -= item_size
        knapsack.append((item_name, item_size, item_value))
    return knapsack, value
```

   
4. While this approach will solve small versions of the problem, as we know the number of permutations grows incredibly fast with the number of items. 
    
   If items \[A, B, C\], say, are the optimal items to insert in the knapsack, we do not really care in which order such items were inserted. So, that solution is equavalent to \[A, C, B\], \[B, A, C\], \[B, C, A\], etc. However, with the current approach ALL equivalent solutions will be visited, which is a significant waste of computation.  
   
   To avid this, we can use the ```combinations``` method still from ```itertools```.

   The combinations of ```M``` objects from a larger collection of ```N``` objects are simply all the different ways of picking ```M``` such objects from the original ```N```, but *disregarding order*. So, if we have a collection of ```N=4``` items, ```(A, B, C, D)```, the valid combinations of ```M=2``` items are ```(A, B)```, ```(A, C)```, ```(A, D)```, ```(B, C)```, ```(B, D)``` and ```(C, D)```. Notice: ```(B, A)``` is not included because it is the same as ```(A, B)```, etc.

    Test the following code:
   
```python
from itertools import combinations
items = ['A','B','C','D']
print(list(combinations(items,1)))
print(list(combinations(items,2)))
print(list(combinations(items,3)))
```
