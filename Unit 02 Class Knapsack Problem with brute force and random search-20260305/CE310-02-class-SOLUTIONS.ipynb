2. Study the following simulator of knapsack filling process which loads the items in the order they are provided until the maximum capacity is exceeded, at which points it returns the items loaded in the knapsack and value of such items.

```python
def fill_knapsack(items, capacity = 7):
    knapsack = []
    value = 0
    remaining_capacity = capacity
    for item_name, item_size, item_value in items:
        if item_size > remaining_capacity: 
            break
        value += item_value
        remaining_capacity -= item_size
        knapsack.append((item_name, item_size, item_value))
    return knapsack, value
```

4. While this approach will solve small versions of the problem, as we know the number of permutations grows incredibly fast with the number of items. 
    
   If items \[A, B, C\], say, are the optimal items to insert in the knapsack, we do not really care in which order such items were inserted. So, that solution is equavalent to \[A, C, B\], \[B, A, C\], \[B, C, A\], etc. However, with the current approach ALL equivalent solutions will be visited, which is a significant waste of computation.  
   
   To avid this, we can use the ```combinations``` method still from ```itertools```.

   The combinations of ```M``` objects from a larger collection of ```N``` objects are simply all the different ways of picking ```M``` such objects from the original ```N```, but *disregarding order*. So, if we have a collection of ```N=4``` items, ```(A, B, C, D)```, the valid combinations of ```M=2``` items are ```(A, B)```, ```(A, C)```, ```(A, D)```, ```(B, C)```, ```(B, D)``` and ```(C, D)```. Notice: ```(B, A)``` is not included because it is the same as ```(A, B)```, etc.

    Test the following code:
```python
from itertools import combinations
items = ['A','B','C','D']
print(list(combinations(items,1)))
print(list(combinations(items,2)))
print(list(combinations(items,3)))
```

````python
from itertools import combinations

items = [('Tennis Ball', 1, 1), # Name, size, value
         ('Mouse', 1, 2),
         ('Sun Glasses', 1, 2),
         ('Mug', 2, 1),
         ('French Press', 2, 1),
         ('Keyboard', 2, 2),
         ('Laptop',	3, 5),
         # New items here
         # ('Monitor',	4, 3),
         # ('Playstation', 3, 4),
         # ('Bicycle', 5, 5),
         # ('Field Guide to GP', 2, 10),
         # ('World Cup Ball', 4, 7)
        ] 


def all_combinations(items):
    for r in range(0, len(items) + 1):
        for combination in combinations(items, r):
            yield combination

max_value = 0
best_knapsack = None

for comb in all_combinations(items):
    knapsack, value = fill_knapsack(comb, capacity = 7)
    if value > max_value:
        max_value = value
        best_knapsack = knapsack
        
print('**** best result', max_value)
print(best_knapsack)
````

````python
from itertools import combinations
from tqdm.auto import tqdm

items = [('Tennis Ball', 1, 1), # Name, size, value
         ('Mouse', 1, 2),
         ('Sun Glasses', 1, 2),
         ('Mug', 2, 1),
         ('French Press', 2, 1),
         ('Keyboard', 2, 2),
         ('Laptop',	3, 5),
         # New items here
         ('Monitor',	4, 3),
         ('Playstation', 3, 4),
         ('Bicycle', 5, 5),
         ('Field Guide to GP', 2, 10),
         ('World Cup Ball', 4, 7)
        ] 

# items += items

max_value = 0
best_knapsack = None

N = len(items)
for M in tqdm(range(1, N+1)):
    for combination in tqdm(combinations(items, M)):
        knapsack, value = fill_knapsack(combination, capacity = 10)
        if value > max_value:
            max_value = value
            best_knapsack = knapsack
        
print('**** best result', max_value)
print(best_knapsack)
````

````python
def mutate_knapsack_loading_plan(plan):
    plan = np.copy(plan)
    mutation_mask = np.random.rand(len(plan)) < 1/5
    plan[mutation_mask] = ~plan[mutation_mask]
    return plan

def fill_knapsack_boolean(knapsack_assignment, capacity = 10):
    items_selected = list(compress(items, knapsack_assignment))
    return fill_knapsack(items_selected, capacity=capacity)

mutated_knapsack = mutate_knapsack_loading_plan(knapsack_assignment)
loaded_knapsack, value = fill_knapsack_boolean(mutated_knapsack, capacity=50)
print('Actually loaded items:')
print_nicely(loaded_knapsack)
print('value in the knapsack:', value)
````
