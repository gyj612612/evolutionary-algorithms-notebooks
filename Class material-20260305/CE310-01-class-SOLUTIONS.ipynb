````python
# Optimised bin packing simulator

def bin_items(items, capacity):
    bins, start_idx, cumsum  = [], 0, 0
    for i, num in enumerate(items):
        if cumsum + num > capacity:
            bins.append(items[start_idx:i])
            start_idx, cumsum = i, 0
        cumsum += num

    if start_idx < len(items):
        bins.append(items[start_idx:])

    return bins
````
