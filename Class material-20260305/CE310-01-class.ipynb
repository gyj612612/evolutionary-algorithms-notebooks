2. Study the following function that bins the items in a list sequentially (i.e., in the original order) and returns the contents of the bins.

```python
def bin_items(items, capacity):
    bins = [[]] # List of bins initialised with an empty bin
    
    for item in items:
        last_bin = bins[-1] # Get last bin
        if item + sum(last_bin) > capacity: 
            # The item does not fit in the last bin: we need a new bin for this item
            bins.append([item])
        else: # The item fits and so we add it to the bin
            last_bin.append(item)

    return bins        
```
