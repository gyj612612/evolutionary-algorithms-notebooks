## Replacement for visualise_tree code 
The code to visualise trees that I provided in the lab has an external dependency (graphviz). 
This is not installed in some machines. So, some students had trouble visualising trees.

Below is a replacement of my visualise_tree code which does not have external dependencies.

```python
# INSTALL nltk AND svgling IF NEEDED

import nltk
from nltk import Tree

def visualise_tree(node):
    # If the node is a list, it's a branch
    if isinstance(node, (list, tuple)):
        label = str(node[0])
        # Convert all remaining items in the list to children
        children = [visualise_tree(child) for child in node[1:]]
        return Tree(label, children)
    # Otherwise, it's a leaf
    else:
        return str(node)

my_program = ['+', 
              ['/', 
               ['-', ['+', ['*', 2, 3], 4], ['+', ['*', 2, 3], 4]], 
               ['-', ['+', ['*', 2, 3], 4], ['+', ['*', 2, 3], 4]]
              ],
              ['/', 
               ['-', ['+', ['*', 2, 3], 4], ['+', ['*', 2, 3], 4]], 
               ['-', ['+', ['*', 2, 3], 4], ['+', ['*', 2, 3], 4]]
              ]
             ]

visualise_tree(my_program)
```
